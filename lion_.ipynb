```python
# ========================= Complete ViT Training with Multiple Optimizers =========================
import os, pickle, numpy as np, matplotlib.pyplot as plt, tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.layers import Dense, Dropout, LayerNormalization, GlobalAveragePooling1D
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, Callback
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import (confusion_matrix, ConfusionMatrixDisplay, classification_report,
                             roc_auc_score, roc_curve, precision_recall_curve, average_precision_score,
                             balanced_accuracy_score, top_k_accuracy_score)
from sklearn.preprocessing import label_binarize
from sklearn.utils.class_weight import compute_class_weight
import pandas as pd

# ----------------------------- PATHS / HYPERPARAMS -----------------------------
TRAIN_DIR = "./train_augmented"
TEST_DIR  = "./augmented_testing_images"
SAVE_DIR  = "./models"; os.makedirs(SAVE_DIR, exist_ok=True)
EXPLAIN_DIR = "./explain"; os.makedirs(EXPLAIN_DIR, exist_ok=True)

IMAGE_SIZE    = (224, 224)
PATCH_SIZE    = 8
EMBED_DIM     = 64
NUM_HEADS     = 4
FF_DIM        = 128
NUM_ENCODERS  = 2
BATCH_SIZE    = 64
EPOCHS        = 70
VAL_SPLIT     = 0.30
SEED          = 1337
USE_MIXED_PRECISION = False

if USE_MIXED_PRECISION:
    tf.keras.mixed_precision.set_global_policy("mixed_float16")

# ----------------------------- OPTIMIZER CLASSES -----------------------------

# ----------------------------- OPTIMIZER CLASSES (drop-in, version-proof) -----------------------------

# ---- helpers to make per-variable slots work across TF/Keras versions
def _var_key(var):
    if hasattr(var, "experimental_ref"):
        return var.experimental_ref()        # modern TF
    if hasattr(var, "_unique_id"):
        return var._unique_id                # some TF builds
    return var.name                          # fallback

def _get_or_make_slot(opt, table, var, slot_name):
    k = _var_key(var)
    if k not in table:
        if hasattr(opt, "add_variable_from_reference"):
            table[k] = opt.add_variable_from_reference(reference_variable=var, name=slot_name)
        else:
            table[k] = opt.add_variable(
                shape=var.shape, dtype=var.dtype, initializer="zeros", name=f"{slot_name}/{var.name}"
            )
    return table[k]


class LAMB(tf.keras.optimizers.Optimizer):
    """Layer-wise Adaptive Moments optimizer for Batch training (LAMB)."""
    def __init__(self, learning_rate=3e-4, weight_decay=1e-4,
                 beta_1=0.9, beta_2=0.999, epsilon=1e-6,
                 clipnorm=1.0, name="LAMB", **kwargs):
        super().__init__(learning_rate=learning_rate, name=name, clipnorm=clipnorm, **kwargs)
        self.weight_decay = weight_decay; self.beta_1 = beta_1; self.beta_2 = beta_2; self.epsilon = epsilon
        self._m = {}; self._v = {}

    def build(self, var_list):
        super().build(var_list)
        for var in var_list:
            _get_or_make_slot(self, self._m, var, "m")
            _get_or_make_slot(self, self._v, var, "v")

    def update_step(self, gradient, variable, learning_rate=None):
        lr = tf.cast(learning_rate if learning_rate is not None else self.learning_rate, variable.dtype)
        if isinstance(gradient, tf.IndexedSlices): gradient = tf.convert_to_tensor(gradient)
        g = tf.cast(gradient, variable.dtype)

        m = _get_or_make_slot(self, self._m, variable, "m")
        v = _get_or_make_slot(self, self._v, variable, "v")

        b1 = tf.cast(self.beta_1, variable.dtype); b2 = tf.cast(self.beta_2, variable.dtype)
        eps = tf.cast(self.epsilon, variable.dtype); step = tf.cast(self.iterations + 1, variable.dtype)

        m_t = b1 * m + (1.0 - b1) * g
        v_t = b2 * v + (1.0 - b2) * tf.square(g)
        m.assign(m_t); v.assign(v_t)

        m_hat = m_t / (1.0 - tf.pow(b1, step))
        v_hat = v_t / (1.0 - tf.pow(b2, step))
        upd   = m_hat / (tf.sqrt(v_hat) + eps)

        if self._use_weight_decay(variable.name):
            upd = upd + tf.cast(self.weight_decay, variable.dtype) * variable

        w_norm = tf.norm(variable); g_norm = tf.norm(upd)
        trust  = tf.where((w_norm > 0.) & (g_norm > 0.), w_norm / (g_norm + eps), tf.cast(1.0, variable.dtype))

        variable.assign_sub(lr * trust * upd)

    def _use_weight_decay(self, var_name):
        n = str(var_name).lower()
        exclude = ['bias','norm','bn','layernorm','pos_emb','positional','gamma','beta']
        return not any(x in n for x in exclude)

    def get_config(self):
        cfg = super().get_config()
        cfg.update(dict(weight_decay=self.weight_decay, beta_1=self.beta_1, beta_2=self.beta_2, epsilon=self.epsilon))
        return cfg


class Lion(tf.keras.optimizers.Optimizer):
    """Lion optimizer (EvoLved Sign Momentum)."""
    def __init__(self, learning_rate=1e-4, beta_1=0.9, beta_2=0.99,
                 weight_decay=1e-4, clipnorm=1.0, name="Lion", **kwargs):
        super().__init__(learning_rate=learning_rate, name=name, clipnorm=clipnorm, **kwargs)
        self.beta_1 = beta_1; self.beta_2 = beta_2; self.weight_decay = weight_decay
        self._m = {}

    def build(self, var_list):
        super().build(var_list)
        for var in var_list:
            _get_or_make_slot(self, self._m, var, "m")

    def update_step(self, gradient, variable, learning_rate=None):
        lr = tf.cast(learning_rate if learning_rate is not None else self.learning_rate, variable.dtype)
        if isinstance(gradient, tf.IndexedSlices): gradient = tf.convert_to_tensor(gradient)
        g = tf.cast(gradient, variable.dtype)

        m = _get_or_make_slot(self, self._m, variable, "m")
        b1 = tf.cast(self.beta_1, variable.dtype); b2 = tf.cast(self.beta_2, variable.dtype)

        upd = tf.sign(b1 * m + (1.0 - b1) * g)

        if self._use_weight_decay(variable.name):
            variable.assign_sub(lr * (upd + tf.cast(self.weight_decay, variable.dtype) * variable))
        else:
            variable.assign_sub(lr * upd)

        m.assign(b2 * m + (1.0 - b2) * g)

    def _use_weight_decay(self, var_name):
        n = str(var_name).lower(); exclude = ['bias','norm','bn','layernorm','pos_emb','gamma','beta']
        return not any(x in n for x in exclude)

    def get_config(self):
        cfg = super().get_config()
        cfg.update(dict(beta_1=self.beta_1, beta_2=self.beta_2, weight_decay=self.weight_decay))
        return cfg


class AdaBelief(tf.keras.optimizers.Optimizer):
    """AdaBelief optimizer."""
    def __init__(self, learning_rate=1e-3, beta_1=0.9, beta_2=0.999,
                 epsilon=1e-16, weight_decay=1e-4, clipnorm=1.0,
                 name="AdaBelief", **kwargs):
        super().__init__(learning_rate=learning_rate, name=name, clipnorm=clipnorm, **kwargs)
        self.beta_1 = beta_1; self.beta_2 = beta_2; self.epsilon = epsilon; self.weight_decay = weight_decay
        self._m = {}; self._s = {}

    def build(self, var_list):
        super().build(var_list)
        for var in var_list:
            _get_or_make_slot(self, self._m, var, "m")
            _get_or_make_slot(self, self._s, var, "s")

    def update_step(self, gradient, variable, learning_rate=None):
        lr = tf.cast(learning_rate if learning_rate is not None else self.learning_rate, variable.dtype)
        if isinstance(gradient, tf.IndexedSlices): gradient = tf.convert_to_tensor(gradient)
        g = tf.cast(gradient, variable.dtype)

        m = _get_or_make_slot(self, self._m, variable, "m")
        s = _get_or_make_slot(self, self._s, variable, "s")

        b1 = tf.cast(self.beta_1, variable.dtype); b2 = tf.cast(self.beta_2, variable.dtype)
        eps = tf.cast(self.epsilon, variable.dtype); step = tf.cast(self.iterations + 1, variable.dtype)

        m_t = b1 * m + (1.0 - b1) * g; m.assign(m_t)
        resid = g - m_t
        s_t = b2 * s + (1.0 - b2) * tf.square(resid) + eps; s.assign(s_t)

        m_hat = m_t / (1.0 - tf.pow(b1, step)); s_hat = s_t / (1.0 - tf.pow(b2, step))
        upd   = m_hat / (tf.sqrt(s_hat) + eps)

        if self._use_weight_decay(variable.name):
            upd = upd + tf.cast(self.weight_decay, variable.dtype) * variable

        variable.assign_sub(lr * upd)

    def _use_weight_decay(self, var_name):
        n = str(var_name).lower(); exclude = ['bias','norm','bn','layernorm','pos_emb','gamma','beta']
        return not any(x in n for x in exclude)

    def get_config(self):
        cfg = super().get_config()
        cfg.update(dict(beta_1=self.beta_1, beta_2=self.beta_2, epsilon=self.epsilon, weight_decay=self.weight_decay))
        return cfg


def get_adamw(lr_schedule):
    """AdamW: Adam with decoupled weight decay."""
    return tf.keras.optimizers.AdamW(
        learning_rate=lr_schedule,
        weight_decay=1e-4,
        beta_1=0.9,
        beta_2=0.999,
        epsilon=1e-6,
        clipnorm=1.0
    )


# ----------------------------- ViT MINI -----------------------------
class PatchEmbedding(layers.Layer):
    def __init__(self, patch_size, embed_dim, **kwargs):
        super().__init__(**kwargs)
        self.proj = layers.Conv2D(embed_dim, kernel_size=patch_size, strides=patch_size, padding="valid", name="patch_proj")
        self.flatten = layers.Reshape((-1, embed_dim))
    def call(self, images):
        conv = self.proj(images)
        tokens = self.flatten(conv)
        tokens._conv_features = conv
        return tokens

class PositionalEmbedding(layers.Layer):
    def __init__(self, num_patches, embed_dim, **kwargs):
        super().__init__(**kwargs)
        self.pos = self.add_weight(name="pos_emb", shape=(1, num_patches, embed_dim), initializer="zeros", trainable=True)
    def call(self, x):
        return x + self.pos

class EncoderBlock(layers.Layer):
    def __init__(self, embed_dim, num_heads, ff_dim, dropout=0.3, **kwargs):
        super().__init__(**kwargs)
        self.norm1 = LayerNormalization(epsilon=1e-6)
        self.attn  = layers.MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim // num_heads)
        self.drop1 = Dropout(dropout)
        self.norm2 = LayerNormalization(epsilon=1e-6)
        self.mlp   = tf.keras.Sequential([Dense(ff_dim, activation="relu"), Dense(embed_dim)])
        self.drop2 = Dropout(dropout)
    def call(self, x, training=False):
        y = self.norm1(x)
        y = self.attn(y, y); y = self.drop1(y, training=training)
        x = x + y
        y = self.norm2(x)
        y = self.mlp(y); y = self.drop2(y, training=training)
        return x + y

# ----------------------------- Data Loading -----------------------------
train_datagen = ImageDataGenerator(
    rescale=1./255, validation_split=VAL_SPLIT,
    horizontal_flip=True, rotation_range=10, width_shift_range=0.05, height_shift_range=0.05, zoom_range=0.1
)
train_gen = train_datagen.flow_from_directory(
    TRAIN_DIR, target_size=IMAGE_SIZE, batch_size=BATCH_SIZE,
    class_mode='categorical', subset='training', shuffle=True, seed=SEED
)
val_gen = train_datagen.flow_from_directory(
    TRAIN_DIR, target_size=IMAGE_SIZE, batch_size=BATCH_SIZE,
    class_mode='categorical', subset='validation', shuffle=False, seed=SEED
)
class_names = [None]*len(train_gen.class_indices)
for name, idx in train_gen.class_indices.items():
    class_names[idx] = name
NUM_CLASSES = len(class_names)
test_datagen = ImageDataGenerator(rescale=1./255)
test_gen = test_datagen.flow_from_directory(
    TEST_DIR, target_size=IMAGE_SIZE, batch_size=BATCH_SIZE, class_mode='categorical', shuffle=False
)

# ----------------------------- Class Weights -----------------------------
y_train = train_gen.classes
class_weights_arr = compute_class_weight(class_weight="balanced", classes=np.arange(NUM_CLASSES), y=y_train)
class_weight = {int(i): float(w) for i, w in enumerate(class_weights_arr)}

# ----------------------------- Learning Rate Schedule -----------------------------
@tf.keras.utils.register_keras_serializable(package="custom")
class WarmupThenCosine(tf.keras.optimizers.schedules.LearningRateSchedule):
    def __init__(self, base_lr, warmup_steps, total_steps, min_lr_ratio=1e-2, name="WarmupThenCosine"):
        super().__init__()
        self.base_lr = float(base_lr)
        self.warmup_steps = int(warmup_steps)
        self.total_steps = int(total_steps)
        self.min_lr_ratio = float(min_lr_ratio)
        self.name = name

        decay_steps = max(1, int(self.total_steps - self.warmup_steps))
        self._cosine = tf.keras.optimizers.schedules.CosineDecay(
            initial_learning_rate=self.base_lr,
            decay_steps=decay_steps,
            alpha=self.min_lr_ratio
        )

    def __call__(self, step):
        step = tf.cast(step, tf.float32)
        warmup_steps = tf.cast(self.warmup_steps, tf.float32)
        base_lr = tf.cast(self.base_lr, tf.float32)

        warm = base_lr * step / tf.maximum(1.0, warmup_steps)
        return tf.where(step < warmup_steps, warm, self._cosine(step - warmup_steps))

    def get_config(self):
        return {
            "base_lr": self.base_lr,
            "warmup_steps": self.warmup_steps,
            "total_steps": self.total_steps,
            "min_lr_ratio": self.min_lr_ratio,
            "name": self.name,
        }

    @classmethod
    def from_config(cls, config):
        return cls(**config)

initial_lr = 3e-4  # Change to 1e-4 for Lion optimizer
steps_per_epoch = max(1, train_gen.samples // BATCH_SIZE)
total_steps  = steps_per_epoch * EPOCHS
warmup_steps = int(0.05 * total_steps)
lr_schedule = WarmupThenCosine(base_lr=initial_lr, warmup_steps=warmup_steps, total_steps=total_steps, min_lr_ratio=1e-2)

class LrLogger(Callback):
    def __init__(self): super().__init__(); self.epochs, self.lrs = [], []
    def _current_lr(self):
        opt = self.model.optimizer; lr = getattr(opt, "learning_rate", None)
        try:
            if isinstance(lr, tf.keras.optimizers.schedules.LearningRateSchedule): lr_t = lr(opt.iterations)
            else: lr_t = lr
            return float(tf.keras.backend.get_value(lr_t))
        except Exception:
            try: return float(tf.keras.backend.get_value(opt.lr))
            except Exception: return float(tf.keras.backend.get_value(lr))
    def on_epoch_end(self, epoch, logs=None): self.epochs.append(epoch+1); self.lrs.append(self._current_lr())
lr_logger = LrLogger()

# ----------------------------- Build Model -----------------------------
def build_model(input_shape, num_classes, lr_schedule, optimizer_name='LAMB'):
    inputs = layers.Input(shape=input_shape)
    patches = PatchEmbedding(PATCH_SIZE, EMBED_DIM, name="patch_embed")(inputs)
    num_patches = (input_shape[0] // PATCH_SIZE) * (input_shape[1] // PATCH_SIZE)
    tokens = PositionalEmbedding(num_patches, EMBED_DIM, name="pos_embed")(patches)
    x = tokens
    for i in range(NUM_ENCODERS):
        x = EncoderBlock(EMBED_DIM, NUM_HEADS, FF_DIM, name=f"enc{i+1}")(x)
    x = GlobalAveragePooling1D(name="token_avg")(x)
    x = Dense(128, activation='relu')(x)
    x = Dropout(0.2)(x)
    out_dtype = "float32" if USE_MIXED_PRECISION else None
    outputs = Dense(num_classes, activation='softmax', dtype=out_dtype, name="pred")(x)
    model = models.Model(inputs, outputs, name="ViT_mini_easy")
    
    # ====== SELECT OPTIMIZER HERE ======
    if optimizer_name == 'LAMB':
        opt = LAMB(learning_rate=lr_schedule, weight_decay=1e-4, clipnorm=1.0)
    elif optimizer_name == 'Lion':
        opt = Lion(learning_rate=lr_schedule, weight_decay=1e-4, clipnorm=1.0)
    elif optimizer_name == 'AdaBelief':
        opt = AdaBelief(learning_rate=lr_schedule, weight_decay=1e-4, clipnorm=1.0)
    elif optimizer_name == 'AdamW':
        opt = get_adamw(lr_schedule)
    else:
        opt = LAMB(learning_rate=lr_schedule, weight_decay=1e-4, clipnorm=1.0)
    
    model.compile(
        optimizer=opt, 
        loss=tf.keras.losses.CategoricalCrossentropy(from_logits=False, label_smoothing=0.1), 
        metrics=['accuracy']
    )
    return model

# ----------------------------- CHOOSE YOUR OPTIMIZER HERE -----------------------------
OPTIMIZER_NAME = 'Lion'  # Options: 'LAMB', 'Lion', 'AdaBelief', 'AdamW'

# If using Lion, uncomment the line below:
initial_lr = 1e-4
lr_schedule = WarmupThenCosine(base_lr=initial_lr, warmup_steps=warmup_steps, total_steps=total_steps, min_lr_ratio=1e-2)

MODEL_PATH = os.path.join(SAVE_DIR, f"{OPTIMIZER_NAME}_model.keras")
HIST_PATH  = os.path.join(SAVE_DIR, f"{OPTIMIZER_NAME}_history.pkl")

model = build_model(IMAGE_SIZE + (3,), NUM_CLASSES, lr_schedule, optimizer_name=OPTIMIZER_NAME)
model.summary(line_length=120)

callbacks = [
    EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True),
    ModelCheckpoint(MODEL_PATH, save_best_only=True),
    lr_logger
]

# ----------------------------- Training -----------------------------
print(f"\n{'='*60}")
print(f"Training with {OPTIMIZER_NAME} optimizer")
print(f"{'='*60}\n")

history = model.fit(
    train_gen, epochs=EPOCHS, validation_data=val_gen,
    callbacks=callbacks, class_weight=class_weight
)

with open(HIST_PATH, "wb") as f:
    pickle.dump(history.history, f)

# ----------------------------- Plot Learning Rate -----------------------------
plt.figure(figsize=(7,4)); plt.plot(lr_logger.epochs, lr_logger.lrs)
plt.title(f"Learning rate per epoch ({OPTIMIZER_NAME})"); plt.xlabel("Epoch"); plt.ylabel("LR")
plt.grid(True, alpha=0.3); plt.tight_layout(); plt.show()

# ----------------------------- Plot Training Curves -----------------------------
hist = history.history
acc_key = "accuracy" if "accuracy" in hist else "acc"
val_acc_key = "val_accuracy" if "val_accuracy" in hist else "val_acc"

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Loss
ax[0].plot(hist["loss"], label='Train Loss')
ax[0].plot(hist["val_loss"], label='Val Loss')
ax[0].set_title(f'Loss over epochs ({OPTIMIZER_NAME})')
ax[0].set_xlabel('Epoch'); ax[0].set_ylabel('Loss')
ax[0].grid(True, alpha=0.3); ax[0].legend()

# Accuracy
ax[1].plot(hist[acc_key], label='Train Acc')
ax[1].plot(hist[val_acc_key], label='Val Acc')
ax[1].set_title(f'Accuracy over epochs ({OPTIMIZER_NAME})')
ax[1].set_xlabel('Epoch'); ax[1].set_ylabel('Accuracy')
ax[1].grid(True, alpha=0.3); ax[1].legend()

plt.tight_layout(); plt.show()

# ----------------------------- Test Evaluation -----------------------------
y_true_onehot = []
for i in range(len(test_gen)):
    _, labels = test_gen[i]; y_true_onehot.append(labels)
y_true_onehot = np.vstack(y_true_onehot)
y_true = np.argmax(y_true_onehot, axis=1)

test_gen.reset()
y_score = model.predict(test_gen, verbose=1)
y_pred = np.argmax(y_score, axis=-1)

overall_acc = np.mean(y_pred == y_true)
print(f"\n{'='*60}")
print(f"{OPTIMIZER_NAME} - Overall Test Accuracy: {overall_acc*100:.2f}%")
print(f"{'='*60}\n")

print("=== Classification Report (TEST) ===")
print(classification_report(y_true, y_pred, target_names=class_names, digits=4, zero_division=0))

# ----------------------------- Confusion Matrix -----------------------------
cm = confusion_matrix(y_true, y_pred, labels=np.arange(NUM_CLASSES))
ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names).plot(cmap=plt.cm.Blues, values_format='d')
plt.title(f'Confusion Matrix ({OPTIMIZER_NAME})'); plt.xticks(rotation=45, ha='right'); plt.tight_layout(); plt.show()

# ----------------------------- Per-Class Metrics -----------------------------
report_dict = classification_report(y_true, y_pred, target_names=class_names, digits=4, zero_division=0, output_dict=True)
per_class_prec = [report_dict[name]["precision"] for name in class_names]
per_class_rec  = [report_dict[name]["recall"]    for name in class_names]
per_class_f1   = [report_dict[name]["f1-score"]  for name in class_names]

x = np.arange(NUM_CLASSES)
plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.bar(x, per_class_prec); plt.title("Precision per class"); plt.xticks(x, class_names, rotation=45, ha="right"); plt.ylim(0,1)
plt.subplot(1,3,2); plt.bar(x, per_class_rec);  plt.title("Recall per class");    plt.xticks(x, class_names, rotation=45, ha="right"); plt.ylim(0,1)
plt.subplot(1,3,3); plt.bar(x, per_class_f1);   plt.title("F1-score per class");  plt.xticks(x, class_names, rotation=45, ha="right"); plt.ylim(0,1)
plt.suptitle(f'{OPTIMIZER_NAME} - Per-Class Metrics')
plt.tight_layout(); plt.show()

print(f"\n{OPTIMIZER_NAME} training and evaluation complete!")
print(f"Model saved to: {MODEL_PATH}")
print(f"History saved to: {HIST_PATH}")
```
