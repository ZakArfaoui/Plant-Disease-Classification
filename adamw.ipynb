```python
# ========================= easy_vit_train_eval.py =========================
# A small, clean Vision-Transformer-like image classifier + evaluation + XAI
# (Grad-CAM on patch-proj, LIME, SHAP)
# --------------------------------------------------------------------------
import os, pickle, numpy as np, matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow.keras import layers, models
from tensorflow.keras.layers import Dense, Dropout, LayerNormalization, GlobalAveragePooling1D
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, Callback
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from sklearn.metrics import (confusion_matrix, ConfusionMatrixDisplay, classification_report,
                             roc_auc_score, roc_curve, precision_recall_curve, average_precision_score,
                             balanced_accuracy_score, top_k_accuracy_score)
from sklearn.preprocessing import label_binarize
from sklearn.utils.class_weight import compute_class_weight

# Optional XAI libs (import lazily later where used)
# pip install lime shap

# -----------------------------
# 0) PATHS — EDIT THESE
# -----------------------------
TRAIN_DIR = "./train_augmented"
TEST_DIR  = "./augmented_testing_images"
SAVE_DIR  = "./models"
os.makedirs(SAVE_DIR, exist_ok=True)
MODEL_PATH = os.path.join(SAVE_DIR, "adamw_model.keras")
HIST_PATH  = os.path.join(SAVE_DIR, "adamw_history.pkl")
EXPLAIN_DIR = "./explain"
os.makedirs(EXPLAIN_DIR, exist_ok=True)

# -----------------------------
# 1) SIMPLE, CLEAR HYPERPARAMS
# -----------------------------
IMAGE_SIZE    = (224, 224)
PATCH_SIZE    = 8   # 16x16 pixel patches
EMBED_DIM     = 64         # channel size inside the transformer
NUM_HEADS     = 4
FF_DIM        = 128        # MLP hidden size inside the encoder
NUM_ENCODERS  = 2
BATCH_SIZE    = 64
EPOCHS        = 50
VAL_SPLIT     = 0.30
SEED          = 1337
SAMPLES_TO_PLOT = 8

# -----------------------------
# 2) MODEL: EASY-TO-UNDERSTAND ViT MINI
# -----------------------------
class PatchEmbedding(layers.Layer):
    """Turns an image into a sequence of patch embeddings via Conv2D(kernel=stride=patch)."""
    def __init__(self, patch_size, embed_dim, **kwargs):
        super().__init__(**kwargs)
        self.patch_size = patch_size
        self.proj = layers.Conv2D(embed_dim, kernel_size=patch_size,
                                  strides=patch_size, padding="valid", name="patch_proj")
        self.flatten = layers.Reshape((-1, embed_dim))  # (B, N_patches, D)

    def call(self, images):
        # We keep both the conv features (H/P, W/P, D) and flattened tokens for Grad-CAM
        conv = self.proj(images)    # (B, H/P, W/P, D)
        tokens = self.flatten(conv) # (B, N, D)
        # Attach conv features to tensor for later retrieval through Keras graph
        tokens._conv_features = conv
        return tokens

class PositionalEmbedding(layers.Layer):
    """Adds a learnable position vector to each patch."""
    def __init__(self, num_patches, embed_dim, **kwargs):
        super().__init__(**kwargs)
        self.pos = self.add_weight(
            name="pos_emb",
            shape=(1, num_patches, embed_dim),
            initializer="zeros",
            trainable=True
        )

    def call(self, x):  # x: (B, N, D)
        return x + self.pos

class EncoderBlock(layers.Layer):
    """One Transformer encoder block (pre-norm)."""
    def __init__(self, embed_dim, num_heads, ff_dim, dropout=0.3, **kwargs):
        super().__init__(**kwargs)
        self.norm1 = LayerNormalization(epsilon=1e-6)
        self.attn  = layers.MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim // num_heads)
        self.drop1 = Dropout(dropout)
        self.norm2 = LayerNormalization(epsilon=1e-6)
        self.mlp   = tf.keras.Sequential([
            Dense(ff_dim, activation="relu"),
            Dense(embed_dim)
        ])
        self.drop2 = Dropout(dropout)

    def call(self, x, training=False):
        y = self.norm1(x)
        y = self.attn(y, y)
        y = self.drop1(y, training=training)
        x = x + y
        y = self.norm2(x)
        y = self.mlp(y)
        y = self.drop2(y, training=training)
        return x + y

def build_model(input_shape, num_classes, lr_schedule):
    """Assemble the full ViT-mini model."""
    inputs = layers.Input(shape=input_shape)

    # Patching + position
    patches = PatchEmbedding(PATCH_SIZE, EMBED_DIM, name="patch_embed")(inputs)
    num_patches = (input_shape[0] // PATCH_SIZE) * (input_shape[1] // PATCH_SIZE)
    tokens = PositionalEmbedding(num_patches, EMBED_DIM, name="pos_embed")(patches)

    # A tiny stack of encoders
    x = tokens
    for i in range(NUM_ENCODERS):
        x = EncoderBlock(EMBED_DIM, NUM_HEADS, FF_DIM, name=f"enc{i+1}")(x)

    # Classification head
    x = GlobalAveragePooling1D(name="token_avg")(x)  # average across patch tokens
    x = Dense(128, activation='relu')(x)
    x = Dropout(0.2)(x)
    outputs = Dense(num_classes, activation='softmax', name="pred")(x)

    model = models.Model(inputs, outputs, name="ViT_mini_easy")

    # AdamW optimizer with cosine schedule
    try:
        from tensorflow.keras.optimizers import AdamW
    except Exception:
        from tensorflow.keras.optimizers.experimental import AdamW
    opt = AdamW(learning_rate=lr_schedule, weight_decay=1e-4)

    # Categorical labels + label smoothing
    model.compile(
        optimizer=opt,
        loss=tf.keras.losses.CategoricalCrossentropy(from_logits=False, label_smoothing=0.1),
        metrics=['accuracy']
    )
    return model

# -----------------------------
# 3) DATA LOADERS
# -----------------------------
train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=VAL_SPLIT,
    horizontal_flip=True,
    rotation_range=10,
    width_shift_range=0.05,
    height_shift_range=0.05,
    zoom_range=0.1
)

train_gen = train_datagen.flow_from_directory(
    TRAIN_DIR, target_size=IMAGE_SIZE, batch_size=BATCH_SIZE,
    class_mode='categorical', subset='training', shuffle=True, seed=SEED
)

val_gen = train_datagen.flow_from_directory(
    TRAIN_DIR, target_size=IMAGE_SIZE, batch_size=BATCH_SIZE,
    class_mode='categorical', subset='validation', shuffle=False, seed=SEED
)

# class names in index order
class_names = [None]*len(train_gen.class_indices)
for name, idx in train_gen.class_indices.items():
    class_names[idx] = name
NUM_CLASSES = len(class_names)
print("[info] Classes:", class_names)

# Keep test set untouched for final evaluation
test_datagen = ImageDataGenerator(rescale=1./255)
test_gen = test_datagen.flow_from_directory(
    TEST_DIR, target_size=IMAGE_SIZE, batch_size=BATCH_SIZE,
    class_mode='categorical', shuffle=False
)

# -----------------------------
# 4) CLASS WEIGHTS (imbalance)
# -----------------------------
y_train = train_gen.classes  # available even with categorical mode
class_weights_arr = compute_class_weight(
    class_weight="balanced",
    classes=np.arange(NUM_CLASSES),
    y=y_train
)
class_weight = {int(i): float(w) for i, w in enumerate(class_weights_arr)}
print("[info] Class weights:", class_weight)

# -----------------------------
# 5) LR SCHEDULE + LOGGER
# -----------------------------
initial_lr = 3e-4
steps_per_epoch = max(1, train_gen.samples // BATCH_SIZE)
first_decay_steps = steps_per_epoch * 10  # ~first restart at epoch 10

lr_schedule = tf.keras.optimizers.schedules.CosineDecayRestarts(
    initial_learning_rate=initial_lr,
    first_decay_steps=first_decay_steps,
    t_mul=2.0, m_mul=1.0, alpha=1e-2
)

class LrLogger(Callback):
    """Record the *actual* LR at each epoch end (robust to TF versions)."""
    def __init__(self):
        super().__init__()
        self.epochs, self.lrs = [], []

    def _current_lr(self):
        opt = self.model.optimizer
        lr = getattr(opt, "learning_rate", None)
        try:
            if isinstance(lr, tf.keras.optimizers.schedules.LearningRateSchedule):
                lr_t = lr(opt.iterations)
            else:
                lr_t = lr
            return float(tf.keras.backend.get_value(lr_t))
        except Exception:
            try:
                return float(tf.keras.backend.get_value(opt.lr))
            except Exception:
                return float(tf.keras.backend.get_value(lr))

    def on_epoch_end(self, epoch, logs=None):
        self.epochs.append(epoch + 1)
        self.lrs.append(self._current_lr())

lr_logger = LrLogger()

# -----------------------------
# 6) BUILD, TRAIN, SAVE
# -----------------------------
model = build_model(IMAGE_SIZE + (3,), NUM_CLASSES, lr_schedule)
model.summary(line_length=120)

callbacks = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    ModelCheckpoint(MODEL_PATH, save_best_only=True),
    lr_logger
]

history = model.fit(
    train_gen,
    epochs=EPOCHS,
    validation_data=val_gen,
    callbacks=callbacks,
    class_weight=class_weight
)

with open(HIST_PATH, "wb") as f:
    pickle.dump(history.history, f)
print(f"[ok] Saved history to {HIST_PATH}")

# Plot LR per epoch
plt.figure(figsize=(7,4))
plt.plot(lr_logger.epochs, lr_logger.lrs, marker='o')
plt.title("Learning rate per epoch (CosineDecayRestarts)")
plt.xlabel("Epoch"); plt.ylabel("LR")
plt.grid(True, alpha=0.3); plt.tight_layout(); plt.show()

# -----------------------------
# 7) TEST EVALUATION
# -----------------------------
# Extract true labels from test generator (categorical → one-hot)
y_true_onehot = []
for i in range(len(test_gen)):
    _, labels = test_gen[i]
    y_true_onehot.append(labels)
y_true_onehot = np.vstack(y_true_onehot)
y_true = np.argmax(y_true_onehot, axis=1)

# Get predictions
test_gen.reset()
y_score = model.predict(test_gen, verbose=1)
y_pred = np.argmax(y_score, axis=-1)

overall_acc = np.mean(y_pred == y_true)
print(f"\nOverall Accuracy (TEST): {overall_acc*100:.2f}%")

# ---- Classification report
print("\n=== Classification Report (TEST) ===")
print(classification_report(y_true, y_pred, target_names=class_names, digits=4, zero_division=0))

# Bar plots for per-class metrics
report_dict = classification_report(
    y_true, y_pred, target_names=class_names, digits=4,
    zero_division=0, output_dict=True
)
per_class_prec = [report_dict[name]["precision"] for name in class_names]
per_class_rec  = [report_dict[name]["recall"]    for name in class_names]
per_class_f1   = [report_dict[name]["f1-score"]  for name in class_names]

x = np.arange(NUM_CLASSES)
plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.bar(x, per_class_prec); plt.title("Precision per class"); plt.xticks(x, class_names, rotation=45, ha="right"); plt.ylim(0,1)
plt.subplot(1,3,2); plt.bar(x, per_class_rec);  plt.title("Recall per class");    plt.xticks(x, class_names, rotation=45, ha="right"); plt.ylim(0,1)
plt.subplot(1,3,3); plt.bar(x, per_class_f1);   plt.title("F1-score per class");  plt.xticks(x, class_names, rotation=45, ha="right"); plt.ylim(0,1)
plt.tight_layout(); plt.show()

# ---- Confusion Matrices
cm = confusion_matrix(y_true, y_pred, labels=np.arange(NUM_CLASSES))

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
disp.plot(cmap=plt.cm.Blues, values_format='d')
plt.title('Confusion Matrix (Counts) — Test'); plt.xticks(rotation=45, ha='right'); plt.tight_layout(); plt.show()

with np.errstate(invalid='ignore', divide='ignore'):
    cm_norm = cm.astype(float) / cm.sum(axis=1, keepdims=True)
    cm_norm = np.nan_to_num(cm_norm)
disp_norm = ConfusionMatrixDisplay(confusion_matrix=cm_norm, display_labels=class_names)
disp_norm.plot(cmap=plt.cm.Blues, values_format='.2f')
plt.title('Confusion Matrix (Row-Normalized) — Test'); plt.xticks(rotation=45, ha='right'); plt.tight_layout(); plt.show()

# ---- ROC–AUC
num_classes = y_score.shape[1]
y_true_bin = label_binarize(y_true, classes=np.arange(num_classes))

if y_true_bin.shape[1] >= 2 and len(np.unique(y_true)) >= 2:
    auc_micro    = roc_auc_score(y_true_bin, y_score, average='micro',    multi_class='ovr')
    auc_macro    = roc_auc_score(y_true_bin, y_score, average='macro',    multi_class='ovr')
    auc_weighted = roc_auc_score(y_true_bin, y_score, average='weighted', multi_class='ovr')
    print(f"\nROC-AUC micro={auc_micro:.4f}  macro={auc_macro:.4f}  weighted={auc_weighted:.4f}")

    plt.figure(figsize=(7,6))
    for i in range(num_classes):
        if y_true_bin[:, i].sum() == 0:
            print(f"[info] Skipping ROC for '{class_names[i]}' (no positives).")
            continue
        fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_score[:, i])
        auc_i = roc_auc_score(y_true_bin[:, i], y_score[:, i])
        plt.plot(fpr, tpr, linewidth=1.6, label=f"{class_names[i]} (AUC={auc_i:.3f})")

    fpr_micro, tpr_micro, _ = roc_curve(y_true_bin.ravel(), y_score.ravel())
    plt.plot(fpr_micro, tpr_micro, linestyle='--', linewidth=2, label=f"micro (AUC={auc_micro:.3f})")

    plt.plot([0,1],[0,1], linestyle=':', linewidth=1)
    plt.title("ROC — All Classes"); plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate")
    plt.xlim(0,1); plt.ylim(0,1); plt.grid(True, alpha=0.3); plt.legend(loc="lower right", fontsize=9)
    plt.tight_layout(); plt.show()
else:
    print("[info] ROC-AUC skipped (need >=2 classes in y_true).")

# ---- Precision–Recall
if y_true_bin.shape[1] >= 2 and len(np.unique(y_true)) >= 2:
    plt.figure(figsize=(7,6))
    for i in range(num_classes):
        if y_true_bin[:, i].sum() == 0:
            continue
        p, r, _ = precision_recall_curve(y_true_bin[:, i], y_score[:, i])
        ap = average_precision_score(y_true_bin[:, i], y_score[:, i])
        plt.step(r, p, where="post", linewidth=1.6, label=f"{class_names[i]} (AP={ap:.3f})")
    plt.xlabel("Recall"); plt.ylabel("Precision")
    plt.title("Precision–Recall — All Classes")
    plt.grid(True, alpha=0.3); plt.legend(loc="lower left", fontsize=9)
    plt.tight_layout(); plt.show()
else:
    print("[info] PR curves skipped (need >=2 classes in y_true).")

# -----------------------------
# 8) SAMPLE IMAGES TRUE vs PRED
# -----------------------------
def load_img_rescaled(path, target_size):
    img = tf.keras.utils.load_img(path, target_size=target_size)
    arr = tf.keras.utils.img_to_array(img) / 255.0
    return img, arr

rng = np.random.default_rng(SEED)
idxs = rng.choice(len(test_gen.filenames),
                  size=min(SAMPLES_TO_PLOT, len(test_gen.filenames)),
                  replace=False)

plt.figure(figsize=(12, 9))
cols = 4
rows = int(np.ceil(len(idxs) / cols))
test_gen.reset()
for i, idx in enumerate(idxs, 1):
    path = test_gen.filepaths[idx]
    true_idx = y_true[idx]
    true_name = class_names[true_idx]

    _, arr = load_img_rescaled(path, IMAGE_SIZE)
    prob = model.predict(arr[None, ...], verbose=0)[0]
    pred_idx = int(np.argmax(prob))
    pred_name = class_names[pred_idx]
    conf = float(prob[pred_idx])

    ok = (pred_idx == true_idx)
    title = f"{'✓' if ok else '✗'} True: {true_name}\nPred: {pred_name} ({conf:.2f})"
    plt.subplot(rows, cols, i)
    plt.imshow(tf.keras.utils.load_img(path, target_size=IMAGE_SIZE))
    plt.axis("off")
    plt.title(title, fontsize=9, color=("green" if ok else "red"))

plt.tight_layout(); plt.show()

print("\n[done] Training + evaluation complete.")
# ====================================================================

# ========================= POST-TRAIN: IMBALANCE-AWARE SUMMARY =========================
# (uses y_true, y_pred, y_score, class_names, NUM_CLASSES defined above)
import pandas as pd

vals, counts = np.unique(y_true, return_counts=True)
support_pct = counts / counts.sum() * 100.0
dist_df = pd.DataFrame({
    "class_idx": vals,
    "class_name": [class_names[i] for i in vals],
    "count": counts,
    "percent": support_pct
}).sort_values("count", ascending=False).reset_index(drop=True)

print("Test-set class distribution:")
print(dist_df)

plt.figure(figsize=(max(8, 0.6*NUM_CLASSES), 4.5))
order = dist_df["class_idx"].to_list()
plt.bar(range(len(order)), dist_df["count"])
plt.xticks(range(len(order)), dist_df["class_name"], rotation=45, ha="right")
plt.title("Test-set class counts"); plt.ylabel("Images")
plt.tight_layout(); plt.show()

bal_acc = balanced_accuracy_score(y_true, y_pred)
print(f"\nOverall Accuracy  : {np.mean(y_true==y_pred):.4f}")
print(f"Balanced Accuracy : {bal_acc:.4f}  (macro recall)")

rep = classification_report(
    y_true, y_pred, target_names=class_names, digits=4, zero_division=0, output_dict=True
)
print("\n=== Macro/Weighted F1 ===")
print(f"Macro F1   : {rep['macro avg']['f1-score']:.4f}")
print(f"Weighted F1: {rep['weighted avg']['f1-score']:.4f}")

per_class_rows = []
for i, name in enumerate(class_names):
    row = rep.get(name, {})
    per_class_rows.append({
        "class": name,
        "precision": row.get("precision", np.nan),
        "recall": row.get("recall", np.nan),
        "f1": row.get("f1-score", np.nan),
        "support": row.get("support", 0)
    })
per_class_df = pd.DataFrame(per_class_rows).sort_values("support", ascending=False)
print("\nPer-class metrics (sorted by support):")
print(per_class_df)

# Top-k
try:
    top3 = top_k_accuracy_score(y_true, y_score, k=3, labels=np.arange(NUM_CLASSES))
    print(f"\nTop-3 Accuracy      : {top3:.4f}")
except Exception:
    print("\n[info] Skipping Top-3 accuracy (needs y_score probabilities).")

# Specificity & balanced acc per class
cm = confusion_matrix(y_true, y_pred, labels=np.arange(NUM_CLASSES))
tn = np.zeros(NUM_CLASSES, dtype=float)
fp = np.zeros(NUM_CLASSES, dtype=float)
fn = np.zeros(NUM_CLASSES, dtype=float)
tp = np.zeros(NUM_CLASSES, dtype=float)

for i in range(NUM_CLASSES):
    tp[i] = cm[i, i]
    fn[i] = cm[i, :].sum() - tp[i]
    fp[i] = cm[:, i].sum() - tp[i]
    tn[i] = cm.sum() - (tp[i] + fn[i] + fp[i])

with np.errstate(divide="ignore", invalid="ignore"):
    recall_i = np.where((tp + fn) > 0, tp / (tp + fn), np.nan)  # TPR
    spec_i   = np.where((tn + fp) > 0, tn / (tn + fp), np.nan)  # TNR
    bal_i    = (recall_i + spec_i) / 2.0

per_class_bacc = pd.DataFrame({
    "class": class_names,
    "recall (TPR)": recall_i,
    "specificity (TNR)": spec_i,
    "balanced_acc": bal_i,
    "support": cm.sum(axis=1)
}).sort_values("balanced_acc", ascending=True)
print("\nPer-class balanced accuracy (low → high):")
print(per_class_bacc)

with np.errstate(invalid='ignore', divide='ignore'):
    cm_norm = cm.astype(float) / cm.sum(axis=1, keepdims=True)
    cm_norm = np.nan_to_num(cm_norm)

fig_w = 1.0 + 0.55*NUM_CLASSES
fig_h = 1.0 + 0.45*NUM_CLASSES
plt.figure(figsize=(fig_w, fig_h))
im = plt.imshow(cm_norm, cmap="Blues", vmin=0, vmax=1)
plt.colorbar(im, fraction=0.046, pad=0.04)
plt.title("Confusion Matrix — Row Normalized")
plt.xlabel("Predicted"); plt.ylabel("True")
plt.xticks(np.arange(NUM_CLASSES), class_names, rotation=45, ha="right")
plt.yticks(np.arange(NUM_CLASSES), class_names)
thresh = 0.20
for i in range(NUM_CLASSES):
    for j in range(NUM_CLASSES):
        val = cm_norm[i, j]
        if val >= thresh:
            plt.text(j, i, f"{val:.2f}", ha="center", va="center",
                     fontsize=8, color=("black" if val < 0.6 else "white"))
plt.tight_layout(); plt.show()
# =============================================================================================
```
